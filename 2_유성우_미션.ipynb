## 최적의 유성우 관측일

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기
 
def predict_best_meteor_shower_viewing(city):
    # 안내 메시지 초기화
    meteor_shower_string = ''
 
    if city not in cities.values:
        meteor_shower_string = f'아쉽지만 {city}에서의 별자리 관측 정보는 제공되지 않습니다.'
        return meteor_shower_string
 
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]
    constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend'] <= latitude), 'constellation'].tolist()
 
    if not constellations_list:
        meteor_shower_string = f'아쉽지만 {city}에서 볼 수 있는 별자리는 없습니다.'
        return meteor_shower_string
   
    meteor_shower_string = f'{city}에서는 별자리 관측이 가능합니다.\n'
       
    # 도시별 관찰 가능한 별자리 iterate
    for constellation in constellations_list:
        # 별자리와 가장 가까운 유성우 조회
        meteor_shower = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'name'].iloc[0]  
 
        # 유성우 관측이 가능한 시작일과 종료일
        meteor_shower_startdate = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'startdate'].iloc[0]
        meteor_shower_enddate = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'enddate'].iloc[0]
 
        # 유성우가 보일 때의 달의 위상 조회
        moon_phases_list = moon_phases[(moon_phases['date'] >= meteor_shower_startdate) & (moon_phases['date'] <= meteor_shower_enddate)]
 
        # 달이 눈에 보이는 첫 날 조회
        best_moon_date = moon_phases_list.loc[moon_phases_list['percentage'].idxmin()]['date']
 
        # 사용자에게 정보전달
        meteor_shower_string += f"{meteor_shower}, {constellation}, {best_moon_date.to_pydatetime():%Y년 %m월 %d일}\n"
         
    return meteor_shower_string
```

```python
# 도시명을 입력히면 별자리 정보를 알려주는 함수 완성하기
def predict_best_meteor_shower_viewing(city):
    # 안내 메시지 초기화
    meteor_shower_string = ''
 
    if city not in cities.values:
        meteor_shower_string = f'아쉽지만 {city}에서의 별자리 관측 정보는 제공되지 않습니다.'
        return meteor_shower_string
 
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]
    new_constellations_list = new_constellations.loc[(new_constellations['latitudestart'] >= latitude) & (new_constellations['latitudeend'] <= latitude), 'constellation'].tolist()
 
    if not new_constellations_list:
        meteor_shower_string = f'아쉽지만 {city}에서 볼 수 있는 별자리는 없습니다.'
        return meteor_shower_string
   
    meteor_shower_string = f'{city}에서는 다음과 같이 별자리 관측이 가능합니다:\n'
       
    # 도시별 관찰 가능한 별자리 iterate
    for constellation in new_constellations_list:
        # 별자리와 가장 가까운 유성우 조회
        new_meteor_shower = new_meteor_showers.loc[new_meteor_showers['radiant'] == constellation, 'name'].iloc[0]  
 
        # 유성우 관측이 가능한 시작일과 종료일
        new_meteor_shower_startdate = new_meteor_showers.loc[new_meteor_showers['radiant'] == constellation, 'startdate'].iloc[0]
        new_meteor_shower_enddate = new_meteor_showers.loc[new_meteor_showers['radiant'] == constellation, 'enddate'].iloc[0]
 
        # 유성우가 보일 때의 달의 위상 조회
        moon_phases_list = moon_phases[(moon_phases['date'] >= meteor_shower_startdate) & (moon_phases['date'] <= meteor_shower_enddate)]
 
        # 달이 눈에 보이는 첫 날 조회
        best_moon_date = moon_phases_list.loc[moon_phases_list['percentage'].idxmin()]['date']
 
        # 사용자에게 정보전달
        meteor_shower_string += f"{new_meteor_shower}를 잘 보려면 {best_moon_date.to_pydatetime():%Y년 %m월 %d일}에 {constellation}자리를 바라보세요. \n"
         
    return meteor_shower_string
```
