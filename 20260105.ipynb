```python
# 업앤다운

import random
answer = random.randint(1,100)
print(answer)

while True:
    #사용자로부터 숫자 입력받기
    #숫자가 정답인지 조건문 작성
    n = int(input("1~100 사이의 정수를 입력하세요"))
    if answer == n:
        print('정답!')
        break
    elif answer < n:
        print('down')
    elif answer > n:
        print('up')
print('게임종료')
```

```python
n = int(input('숫자:'))
star_count = 0

for i in range(1,11):
    star_remain = n - star_count
    if star_remain > i:
        print('*'*i)
        star_count += i
    else:
        print("*"*star_remain)
        break
```
