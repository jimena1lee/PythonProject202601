```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
```

```python
array = [2,3,4,5,6,7,8,9]
result = list(map(is_prime,array))
print(result)
```

```python
def rank(scores):
    totals = [sum(s) for s in scores]

    # 총점을 내림차순 정렬 (중복 유지)
    sorted_totals = sorted(totals, reverse=True)

    result = []
    for t in totals:
        # 처음 등장하는 위치 + 1 = 등수
        result.append(sorted_totals.index(t) + 1)

    return result
```

```python
score1 = [[80, 70], [90, 50], [40, 70], [50, 80]]
rank(score1) # [1, 2, 4, 3] 
```

```python
score2 = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]
rank(score2) # [4, 4, 6, 2, 2, 1, 7]
```

##### 나혼자복습

```python
# 가위바위보 만들기
# 0: 가위, 1: 바위, 2: 보

import random

rsp = ['가위', '바위', '보']

user = int(input('가위(0), 바위(1), 보(2) 중 하나를 선택하세요: '))
computer = random.randint(0, 2)

print('사용자:', rsp[user])
print('컴퓨터:', rsp[computer])

if user == computer:
    print('비겼습니다')
elif (user == 0 and computer == 2) or \
     (user == 1 and computer == 0) or \
     (user == 2 and computer == 1):
    print('이겼습니다')
else:
    print('졌습니다')
```
