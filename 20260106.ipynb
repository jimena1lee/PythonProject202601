```python
# 미션 "전화번호 검색 프로그램 만들기"
## 메뉴천택 [1:추가, 2:검색, 0:종료]

addr = {
    '김':'010-1111-1111','이':'010-2222-2222',
    '삼':'010-3333-3333','사':'010-4444-4444'}

while True:
    n = input('원하는 서비스를 입력하세요 [1:추가, 2:검색, 0:종료]')

    if n == '1':
        a = input('이름을 입력하세요:')
        b = input('번호를 입력하세요:')
        addr[a] = b
        print('저장완료')


    elif n == '2':
        print(list(addr.keys()))
        c = input('원하는 이름을 작성하세요')
        print(addr.get(c, '등록되지 않은 이름입니다.'))


    elif n == '0':
        print('프로그램을 종료합니다.')
        break
    
    else:
        print('메뉴 선택 오류입니다. 다시 선택하세요.')
        continue
```

```python
#도전 미션 - 모스부호 해석기

morse = { 
    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f', 
    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l', 
    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r', 
    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x', 
    '-.--':'y','--..':'z' 
}

# ".... . .-.. .-.. ---"  "hello" 
# ".--. -.-- - .... --- -."   "python" 


letters = ".... . .-.. .-.. ---" 

letters_list = letters.split()

answer = ''
for i in letters_list:
    answer += morse[i]

answer
```

```python
# 답 찾기

def max_num(a,b,c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val
```

```python
max_num(-3,-5,-1)
```

```python
def num_min(num_list):
    num_list =[1,2,3,4]
    minval = num_list[0]
    if minval > num_list[1]:
        minval = num_list[1]
    if minval > num_list[2]:
        minval = num_list[2]
    if minval > num_list[3]:
        minval = num_list[3]
    return minval
```

```python
num_min([-8,-40,1,2,3,4,5,6,7])
```

```python
#답찾기
def min_num(array):
    min_val = array[0]
    for i in array:
        if i < min_val:
            min_val = i
    return min_val
```

```python
min_num([3,4,5,6,2,1,-80])
```

```python
#윤년 확인하는 함수 만들기

def isLeapYear(n):
    year_result =''
    if n % 4 == 0:
        year_result = True
    if n % 4 == 0 and n % 100 == 0:
        year_result = False
    if n % 400 == 0:
        year_result = True
    else:
        year_result = False
    return year_result
    
```

```python
isLeapYear(2404)
```

```python
#답찾기

def isleapyear(year):
    return (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0)
```

```python
def add(array1):
    total = 0
    for i in array1:
        total += i
    return total

def mul(array2):
    total = 0
    for i in array2:
        total *= i
    return total

def add_mul(array3):
    
    if array3[0] == 'add':
        array3[1:] = array1
        add(array1)

    if array3[0] == 'mul':
        array3[1:] = array2
        mul(array2)
```

```python
add_mul('mul',1,2,3,4,5)
```
